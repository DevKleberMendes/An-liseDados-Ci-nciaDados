```python
# Analise de Renda Per Capita - Brasil (indicador socioeconômicos que avalia o grau de desenvolvimento econômico de um determinado lugar. A média é obtida através da divisão do Produto Nacional Bruto (PNB) pelo número total de habitantes. Dados: http://www.atlasbrasil.org.br/
```

```python
# Bibliotecas
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lendo a base de dados
Base_Dados = pd.read_excel('Dados_Pib.xlsx')

# Verificando
Base_Dados.head(10)
```

```python
Base_Dados.groupby( by=['Territorialidades', 'Ano'] ).mean()
```

```python
# Cor Fundo de Fundo
Cor_Fundo = '#f5f5f5'

# Criar o sistema de Grids
Grid_Graficos = sns.FacetGrid( Base_Dados, col='Territorialidades', hue='Territorialidades', col_wrap=4 )

# Adicionar grafico linhas em cada gráfico
Grid_Graficos = Grid_Graficos.map( plt.plot, 'Ano', 'PIB per capita')

# Adiconar uma sombra + Ajuste do titulo
Grid_Graficos = Grid_Graficos.map( plt.fill_between,  'Ano', 'PIB per capita', alpha=0.2).set_titles('{col_name} Territorialidades')

# Filtrar o titulo
Grid_Graficos = Grid_Graficos.set_titles('{col_name}')

# Adicionar um subtitulo
Grid_Graficos = Grid_Graficos.fig.suptitle(
    'Renda per capita por Estado \n Relatório disponível em nosso GitHub',
    fontsize=18
)

# Ajustando
plt.subplots_adjust( top=0.92 )
```
